## 1. Utiliza sympy y encuentra los valores de las incógnitas que hagan que los siguientes sistemas de ecuaciones sean verdaderos

```python
import sympy as sp

def f(x, y):
    return x + y - 10

def g(x, y):
    return x - y - 2

sp.var('x')
sp.var('y')
R = sp.solve((f(x, y), g(x, y)), (x, y))
R
```

```python
import sympy as sp

def f(x, y):
    return (3*x - 2*y)/3 + 4*y - 13/3

def g(x, y):
    return (-4*y + 2*x)/3 - (3*x)/2 + 13/6

sp.var('x')
sp.var('y')
R = sp.solve((f(x, y), g(x, y)), (x, y))
R
```

```python
import sympy as sp

def f(x, y):
    return x**2 - 2*y - 16

def g(x, y):
    return x - y - 3
    
sp.var('x')
sp.var('y')
R = sp.solve((f(x, y), g(x, y)), (x, y))
R
```

```python
import sympy as sp

def f(x, y, z):
    return x + y + z - 10

def g(x, y):
    return x - y - 2

def a(y, z):
    return y + z + 2
    
sp.var('x')
sp.var('y')
sp.var('z')
R = sp.solve((f(x, y, z), g(x, y), a(y, z)), (x, y, z))
R
```

## 2. Encuentra los puntos críticos usando cálculo diferencial, decide si son máximos o mínimos usando el criterio de la segunda derivada, y grafica para decidir se son óptimos locales o globales:

```python
import matplotlib.pyplot as plt
import sympy as sp

sp.var("x")
def f(x):
    return 1235 + 113*x - 6*x**2

dx = sp.diff(f(x), x) 
pc = sp.solve(dx, x)
print("El punto crítico está en " + str(pc))

Ddx = sp.diff(dx, x)
a = Ddx.subs(x, pc[0])
print(a)

import numpy as np
x = np.linspace(-100,119,100)
y = 1235 + 113*x - 6*x**2
plt.figure()
plt.plot(x,y)
plt.grid()

print('La segunda derivada de f(x) es -12 como es valor negativo significa que es un máximo y al ver la gráfica puedo determinar que es un máximo global')
```

```python
import matplotlib.pyplot as plt
import sympy as sp
import math as m


def f(x):
    return -3 + x + m.pi/x


sp.var("x")
dx = sp.diff(f(x), x) 
pc = sp.solve(dx, x)
print("El punto crítico está en " + str(pc[1]))

Ddx = sp.diff(dx, x)
a = Ddx.subs(x, pc[0])
print(a)

import numpy as np
x = np.linspace(0,20,100)
y = -3 + x + m.pi/x
plt.figure()
plt.plot(x,y)
plt.grid()

print('La segunda derivada de f(x) es negativa lo que significa que es un máximo y al ver la gráfica puedo determinar que es un máximo local')
```
